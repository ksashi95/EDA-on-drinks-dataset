#### 6.How many countries are represented in the dataset?

```python
num_of_countries = df['country'].nunique()
print("number of countries in the dataset: ",num_of_countries)
```

#### 7.What is the distribution of beer servings across countries?

```python
sns.histplot(x = 'beer_servings', data = df)
plt.xlabel('beer_servings')
plt.ylabel('Count')
plt.title('distribution of beer servings')

plt.show()
```

#### 8.Which country has the highest beer servings?

```python
country_highest_beer = df.loc[df['beer_servings'].idxmax(), 'country']

print("\nCountry with Highest Beer Servings:", country_highest_beer)
```

#### 9.What is the distribution of spirit servings across countries?

```python
sns.histplot(x = 'spirit_servings', data = df)
plt.xlabel('spirit_servings')
plt.ylabel('Count')
plt.title('distribution of spirit servings')

plt.show()
```

#### 10.Which country has the highest spirit servings?

```python
country_highest_spirit = df.loc[df['spirit_servings'].idxmax(), 'country']

print("\nCountry with Highest Spirit Servings:", country_highest_spirit)
```

#### 11.What is the distribution of wine servings across countries?

```python
sns.histplot(x = 'wine_servings', data = df)
plt.xlabel('wine_servings')
plt.ylabel('Count')
plt.title('distribution of wine servings')

plt.show()
```

#### 12.Which country has the highest wine servings?

```python
country_highest_wine = df.loc[df['wine_servings'].idxmax(), 'country']

print("\nCountry with Highest Wine Servings:", country_highest_wine)
```

#### 13.How are beer, spirit, and wine servings correlated with each other?

```python
correlation = df[['beer_servings', 'spirit_servings', 'wine_servings']].corr()

print("\nCorrelation between Servings:")
correlation
```

```python
sns.heatmap(correlation.corr(), annot = True)
plt.title('Correlation between Servings')
plt.show()
```

#### 14.What is the total litres of pure alcohol consumed on average per country?

```python
avg_alcohol_consumption = round(df['total_litres_of_pure_alcohol'].mean(),2)

print("\nAverage Total Litres of Pure Alcohol:", avg_alcohol_consumption)
```

#### 15.Is there any relationship between total litres of pure alcohol and beer, spirit, or wine servings?

```python
alcohol_servings_corr = df[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].corr()
alcohol_servings_corr
```

```python
sns.heatmap(alcohol_servings_corr.corr(), annot = True)
plt.title('Relationship between total litres of pure alcohol and beer, spirit, or wine servings')
plt.show()
```

#### 16.Are there any outliers in the dataset? If yes, which countries do they belong to?

```python
outliers = df[df['total_litres_of_pure_alcohol'] > df['total_litres_of_pure_alcohol'].quantile(0.95)]

print("\nOutliers (Total Litres of Pure Alcohol > 95th percentile):")
outliers
```
